```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

x = np.linspace(-4, 4, 1000)
dfs = [5, 10, 30]

for df in dfs:
    plt.plot(x, t.pdf(x, df), label=f'df = {df}')
    mean, var, skew, kurt = t.stats(df, moments='mvsk')
    print(f"Mean: {mean}, Variance: {var}, Skewness: {skew}, Kurtosis: {kurt}")

# Plot standard normal distribution for comparison
plt.plot(x, norm.pdf(x), 'k--', label='Normal Distribution')

plt.title("Student's t-Distribution for Different Degrees of Freedom")
plt.xlabel('t value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()


```
