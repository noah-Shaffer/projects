```python
import pandas as pd

# For Windows
#df = pd.read_csv(r'C:\Users\nshaffer\Desktop\returns_df.csv')

# For Mac/Linux
df = pd.read_csv('/Users/nshaffer/Desktop/returns_df2.csv',index_col=0)

# Remove rows where any value is greater than 1
#df = df[(df <= 1).all(axis=1)]
df.drop(df[df.max(axis=1) > 1].index, inplace=True)
df.drop(df[df.min(axis=1) < -1].index, inplace=True)
df
```

```python
import matplotlib.pyplot as plt

# returns_df: rows = stocks, columns = days after split

for day in df.columns:
    plt.figure(figsize=(7, 4))
    plt.hist(df[day].dropna(), bins=30, edgecolor='black')
    plt.title(f"Histogram of Returns: Day {day} After Reverse Split")
    plt.xlabel("Daily Return")
    plt.ylabel("Frequency")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```
